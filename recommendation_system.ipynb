```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df1['genres'] = df1['genres'].apply(convert)
```

```python
def convert1(obj):
    L = []
    count=0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count+=1
        else:
          break

    return L
```

```python
def extract(val):
  L=[]
  for i in ast.literal_eval(val):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
df1['crew']=df1['crew'].apply(extract)
```

```python
df1['overview']=df1['overview'].apply(lambda x:x.split())
```

```python
df1
```

```python
df1['cast'] = df1['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df1['crew'] = df1['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df1['tags']=df1['keywords']+df1['genres']+df1['cast']+df1['crew']+df1['overview']
df1['tags']
```

```python
new_df1=df1[['id','title','tags']]
```

```python
column_type = new_df1['tags'].dtypes
print(column_type)
```

```python
new_df1['tags'] = new_df1['tags'].apply(lambda x: " ".join(map(str, x)) if isinstance(x, (list, tuple, set)) else str(x))
```

```python
new_df1['tags'] = new_df1['tags'].apply(lambda x:x.lower())
```

```python
new_df1['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df1['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def  convert2(val):
  L=[]
  for i in val.split():
    L.append(ps.stem(i))
  return " ".join(L)
  #we use this converstion to remove the repeating values like eg:dancing, dance are quite similar it will return danc in the output
```

```python
convert2('ulture clash Action SamWorthington ZoeSaldana SigourneyWeaver JamesCameron In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization')
```

```python
new_df1['tags']=new_df1['tags'].apply(convert2)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar=cosine_similarity(vectors)
```

```python
similar[0]
```

```python
def recommend(movie):
    movie_index = new_df1[new_df1['title'] == movie].index[0]
    distances = similar[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df1.iloc[i[0]].title)
```

```python
recommend('Batman')
```
